```python
#                                                    Analysis

#            MOST SELLING PIZZA CATEGORY
top_category = df.groupby('pizza_category').sum().sort_values('quantity',ascending=False).head()

# MOST SELLING PIZZA CATEGORY

# Topcategory=temp.groupby("pizza_category")["order_id"].agg(["count"]).reset_index().sort_values(by="count",ascending=False).head()
# print(Topcategory)
# plt.figure(figsize=(5,5))
# graph=sns.barplot(x='pizza_category',y='count',data=Topcategory)
# graph.set_title('MOST SELLING CATEGORY')
# plt.xticks(rotation=60)
# plt.show()

pd.crosstab(df.pizza_category, df.quantity).plot(kind='bar')
```
