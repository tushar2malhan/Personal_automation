```python
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray() 
for i in range(4):
    plt.matshow(digits.images[i]) 
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
df
# digits.data is mapped with digits.target by default
```

```python
X = df.drop('target',axis='columns')
y = df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=30)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```

```python
# Confusion Matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, model.predict(X_test))
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
# 0 = my model predicted 34 times it (Truth) , one time wrong it predicted 2 == wrong
# 1 = model predicted it 38 times it, one time wronf where it predicted 8
# for number 2 = it predicted 33 times where it was 100 % accurate
```
