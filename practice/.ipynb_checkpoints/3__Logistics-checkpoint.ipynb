```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Logistics HR_comma_sep.csv')

#  Q Average numbers for all columns who legt and who are retained 

df.groupby('left').mean()
#   == Satisfaction level seems to be relatively low (0.44) 
#      in employees leaving the firm vs the retained ones (0.66)


#   Q Impact of salary on employee retention

# pd.crosstab(df.salary,df.left).plot(kind='bar')
#  == Above bar chart shows employees with high salaries are likely to not leave the company

# pd.crosstab(df.Department,df.left).plot(kind='bar')


# From above chart there seem to be some impact of department on employee retention 
# but it is not major hence we will ignore department in our analysis

# From the data analysis so far we can conclude that we will use 
# following variables as independant variables in our model

#  where X are using these factors :-

#          **Satisfaction Level**
#          **Average Monthly Hours**
#          **Promotion Last 5 Years**
#          **Salary**
df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary','left']]
```

```python

#                                 Pre Processing 


subdf = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
# subdf.head()

# - convert string values to numeric 
# - if not possible then create dummies values for the same
# - Dummie values for  Salary

dummies_salary = pd.get_dummies(subdf.salary)
# dummies_salary

df_with_dummies = pd.concat([subdf,dummies_salary],axis='columns')
df_with_dummies

# - Checked == No Value is Null 
# df.isnull().values  # .any()

# Now we need to remove salary column which is text data. 
#  It is already replaced by dummy variables so we can safely remove it

df_with_dummies.drop('salary',axis='columns',inplace=True)
df_with_dummies.head()
```

```python
#                   Finally we compute our X = input   and   y output Values
# from matplotlib import pyplot as plt
# %matplotlib inline
# plt.scatter(final.satisfaction_level, final.low, marker='+',color='red')
X = df_with_dummies
X.head()

y = df.left
y

#                   Lets split the data in train and test 
# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(,train_size=0.8)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
X_test
```
