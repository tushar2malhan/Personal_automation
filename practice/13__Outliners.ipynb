```python
# remove outliers using percentile based on price per night for a given apartment/home. 
# You can use suitable upper and lower limits on percentile based on your intuition. 
# Your goal is to come up with new pandas dataframe that doesn't have the outliers present in it.
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('Outliners Newyork.csv')
df.head()
```

```python
# Pre processsing

# df.isnull().sum()    # multiple columns null
```

```python
df.describe()
# After seeing the difference, we can see " price and minimum_nights " columns have huge numeric gap 
# THis could lead to intolerance of data predictions , so lets use outliners

df[df.price > df.price.quantile(0.95)]   # these include list of rows for  prices more than required 
df[df.price < df.price.quantile(0.05)]   # these include list of rows for  prices less than required 

min_thresold, max_thresold = df.price.quantile( [0.05, 0.95] )

# Lets filter the both so that we can exclude
df2 = df[  (df.price < max_thresold) & ( df.price > min_thresold ) ]

df2.describe()  # Now prices are in stable position
```

```python
# Same way lets remove unwanted rows not having adequate values for column minimum_nights
# df2.describe()
min_thresold, max_thresold = df2.minimum_nights.quantile([0.05,0.95])
final_df = df2[( df2.minimum_nights < max_thresold ) & ( df2.minimum_nights > min_thresold )  ]
final_df.describe()   # we got the adequate values 
```
